```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
L = 0.02
number = 100
points = int(number ** 0.5)
x = np.linspace(0, L, points)
y = np.linspace(0, L, points)
npgrid=np.meshgrid(x,y)
annot = []
for i in range(1, number+1):
    annot.append('N{}'.format(i))

plt.figure(figsize=(10, 10))
plt.scatter(npgrid[0], npgrid[1])
for i in range(number):
    plt.annotate(annot[i], (npgrid[0].flatten()[i]+0.0003, npgrid[1].flatten()[i]-0.0001))
```
